```python
import os
import pandas as pd
```

```python
#This notebook reads in the downloaded csv files and combines them into one for Tableau
directory = 'Citi_Data'
final_csv = 'combined_data.csv'
files = []

#Reading in all of the files in the Resources Directory and adds them to a list
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    print(f)
    # checking if it is a file
    if os.path.isfile(f) and f != os.path.join(directory, final_csv):
        files.append(f)
files
```

```python
df_list = []
#Creating a list of dataframes that are created from the csv files
for file in files:
    df_list.append(pd.read_csv(file))
df_list[-1]
```

```python
#Concatanating the dataframes together, using the 1st dataframe as a base
combined_df = df_list[0]

for n in range(1, len(df_list)):
    combined_df = pd.concat([combined_df, df_list[n]], ignore_index=True, sort=False)
combined_df
```

```python
#Saving csv
combined_df.to_csv(os.path.join(directory, final_csv))
```
